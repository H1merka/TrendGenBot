###BNB 4-bit quantized model loading

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "OpenGVLab/InternVL2_5-4B"

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    load_in_4bit=True,
    trust_remote_code=True
).eval()
```

###Initialization of functions for preprocessing

```python
import torchvision.transforms as T
from torchvision.transforms.functional import InterpolationMode
from PIL import Image
```

```python
# ImageNet Normalization
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD = (0.229, 0.224, 0.225)
```

```python
def build_transform(input_size=448):
    '''Pipeline for transformation'''
    return T.Compose([
        T.Resize((input_size, input_size), interpolation=InterpolationMode.BICUBIC),
        T.ToTensor(),
        T.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
    ])

def preprocess_image(image: Image.Image, image_size=448):
    '''Image transformation'''
    image = image.convert('RGB')
    transform = build_transform(image_size)
    pixel_values = transform(image).unsqueeze(0)
    return pixel_values
```

###Single inference

```python
import requests
import matplotlib.pyplot as plt
```

```python
image_url = "https://www.ilankelman.org/stopsigns/australia.jpg"
generation_config = dict(max_new_tokens=128, do_sample=False)

# Image loading
image = Image.open(requests.get(image_url, stream=True).raw).convert('RGB')

# Image display
plt.imshow(image)
plt.axis('off')
plt.show()

# Preprocessing
pixel_values = preprocess_image(image).to(dtype=torch.bfloat16, device=model.device)

# Text input to model
additional_text = "urban life in Beijing"

# Prompt
question = (
    "<image>\n"
    "Please describe the visual content of the image, "
    "and provide a short paragraph that connects the image to "
    f"the following text concept: '{additional_text}'."
)

# Model inference
response = model.chat(tokenizer, pixel_values, question, generation_config)

print("Question:\n" + question.replace('<image>', '[IMAGE]'))
print("Answer:\n" + response)
```

###Multiple inference

```python
data = [
    ("https://www.ilankelman.org/stopsigns/australia.jpg", "urban life in Beijing"),
    ("https://upload.wikimedia.org/wikipedia/commons/1/1e/Traffic_congestion_Jakarta_2021.jpg", "traffic in Jakarta"),
    ("https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Tiananmen_Square_2023.jpg/1920px-Tiananmen_Square_2023.jpg", "daily routine in Chinese cities")
]
generation_config = dict(max_new_tokens=128, do_sample=False)

all_responses = []

# A loop for processing the list
for image_url, additional_text in data:
    image = Image.open(requests.get(image_url, stream=True).raw).convert('RGB')

    plt.imshow(image)
    plt.axis('off')
    plt.show()

    pixel_values = preprocess_image(image).to(dtype=torch.bfloat16, device=model.device)

    question = (
        "<image>\n"
        "Please describe the visual content of the image, "
        "and provide a short paragraph that connects the image to "
        f"the following text concept: '{additional_text}'."
    )

    response = model.chat(tokenizer, pixel_values, question, generation_config)

    print("Question:\n" + question.replace('<image>', '[IMAGE]'))
    print("Answer:\n" + response)

    all_responses.append(response)

print(all_responses)
```
